```python
column_rename_map = {
  "sal_indicador": "Indicador",
  "sal_codigo": "Codigo",
  "sal_destinoint": "Destino Internacional",
  "sal_valor": "Valor_salida"
}
```

```python
aeropuerto_df_renombrado = aeropuerto_df.rename(columns=column_rename_map)
```

```python
aeropuerto_df_renombrado.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(18, 10))
frecuencia_paises = aeropuerto_df_renombrado['Codigo'].value_counts()

ax = frecuencia_paises.plot(kind='bar', color='skyblue', edgecolor='black')
umbral = 50

for p in ax.patches:
    if p.get_height() > umbral:
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points', rotation=45)
plt.xticks(rotation=45, ha='right')

plt.title('Frecuencia de Codigo de País')

plt.xlabel('Código de País')

plt.ylabel('Frecuencia')

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

aeropuerto_df['sal_periodo'] = pd.to_datetime(aeropuerto_df['sal_periodo'], format='%Y%m')

monthly_trend = aeropuerto_df.groupby('sal_periodo')['sal_valor'].sum().reset_index()

monthly_trend = monthly_trend.sort_values('sal_periodo')

plt.figure(figsize=(15, 7))
plt.plot(monthly_trend['sal_periodo'], monthly_trend['sal_valor'], color='skyblue', linewidth=2)
plt.fill_between(monthly_trend['sal_periodo'], monthly_trend['sal_valor'], color='violet', alpha=0.4)
plt.plot(monthly_trend['sal_periodo'], monthly_trend['sal_valor'], color='black')

plt.title('Tendencia Cronológica de Viajes (Salidas Internacionales)')
plt.xlabel('Fecha')
plt.ylabel('Valor de Salida')
plt.grid(True)
plt.show()
```

```python
us_data = aeropuerto_df_renombrado[aeropuerto_df_renombrado['Codigo'] == 'US'].copy()
us_data.loc[:, 'periodo'] = pd.to_datetime(us_data['sal_periodo'], format='%Y%m')
us_data['año'] = us_data['periodo'].dt.year
us_data['mes'] = us_data['periodo'].dt.month
mejor_mes_por_año = us_data.groupby(['año', 'mes'])['Valor_salida'].sum().reset_index()
mejor_mes_por_año = mejor_mes_por_año.loc[mejor_mes_por_año.groupby('año')['Valor_salida'].idxmax()]
print(mejor_mes_por_año)

plt.figure(figsize=(14, 8))
plt.plot(mejor_mes_por_año['año'].astype(str) + '-' + mejor_mes_por_año['mes'].astype(str),
         mejor_mes_por_año['Valor_salida'], marker='o', linestyle='-', color='b')
plt.title('Mejor mes de cada año para US en términos de Valor_salida')
plt.xlabel('Año-Mes')
plt.ylabel('Valor de Salida')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


import matplotlib.pyplot as plt

total_value_per_destination = aeropuerto_df_renombrado.groupby('Destino Internacional')['Valor_salida'].sum()


top_6_destinations = total_value_per_destination.sort_values(ascending=False).head(6)

print("Ranking de los 6 países/destinos más visitados (en términos de valor de salida):")
print(top_6_destinations)


plt.figure(figsize=(12, 7))
top_6_destinations.plot(kind='bar', color='brown', edgecolor='black')
plt.title('Top 6 Países/Destinos más visitados (Valor de Salida)')
plt.xlabel('Destino Internacional')
plt.ylabel('Valor Total de Salida')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nComparación 1 a 1 de los top 6 destinos (total valor de salida):")
top_6_values = top_6_destinations.to_dict()

destinations = list(top_6_values.keys())

for i in range(len(destinations)):
    for j in range(i + 1, len(destinations)):
        dest1 = destinations[i]
        dest2 = destinations[j]
        value1 = top_6_values[dest1]
        value2 = top_6_values[dest2]
        print(f"{dest1} ({value1:.0f}) vs {dest2} ({value2:.0f})")
        if value1 > value2:
            print(f"  -> {dest1} es mayor.")
        elif value2 > value1:
            print(f"  -> {dest2} es mayor.")
        else:
            print(f"  -> Tienen el mismo valor.")
```
